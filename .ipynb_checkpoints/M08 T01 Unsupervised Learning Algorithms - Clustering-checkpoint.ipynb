### Dimensionality reduction using PCA (Principal Component Analysis)

```python
### Feature Visualization from sklearn.decomposition import PCA
from sklearn.decomposition import PCA
pca = PCA()
_ = pca.fit_transform(fb_df)
PC_components = np.arange(pca.n_components_) + 1
```

```python
# Let's take a look to the Scree Plot, so we can check how many Principal Component need to go on with the clustering
# process study
_ = sns.set(style='whitegrid', font_scale=1.2)
fig, ax = plt.subplots(figsize=(10, 7))
_ = sns.barplot(x=PC_components, y=pca.explained_variance_ratio_, color='b')
_ = sns.lineplot(x=PC_components-1, y=np.cumsum(pca.explained_variance_ratio_), color='black', linestyle='-', linewidth=2, marker='o', markersize=8)

plt.title('Scree Plot')
plt.xlabel('N-th Principal Component')
plt.ylabel('Variance Explained')
plt.ylim(0, 1.2)
plt.show()
```

As shown in the graph we only need PC1 and PC2 to capture more than 90% of the variance.

```python
# Feature Weight
fig, ax = plt.subplots(figsize=(14, 10))
_ = sns.heatmap(pca.components_**2,
                 yticklabels=["PCA"+str(x) for x in range(1,pca.n_components_+1)],
                 xticklabels=fb_df.columns.tolist(),
                 annot=True,
                 fmt='.2f',
                 square=True,
                 linewidths=0.05,
                 cbar_kws={"orientation": "vertical"})
```

```python
# Perform PCA (Choose n_components to keep)
pca = PCA(n_components=2)
pca_array = pca.fit_transform(fb_df)

# Convert back to DataFrame for readability
df_pca = pd.DataFrame(data=pca_array)
df_pca.columns = ['PC' + str(col+1) for col in df_pca.columns.values]
df_pca['label'] = y

df_pca.head()
```

### K-means Clustering

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=42) 

kmeans.fit(df_pca)

labels = kmeans.labels_
```

## Exercise 2: Classify the dataset using of Hierarchical Clustering Algorithm

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize = (10, 7))  
plt.title("Dendrogram")  
shc.dendrogram(shc.linkage(df_pca, method = "ward"))
plt.show()
```

## Exercise 3: Calculate the performance of the clustering using a parameter such as Elbow Method and Silhouette

```python
from kneed import KneeLocator
from sklearn.metrics import silhouette_score
```

```python
# Elbow Method 
# Running several k-means, increment k with each iteration, and record the SSE. 

kmeans_kwargs = {
    "init": "random",
    "n_init": 10,
    "max_iter": 300,
    "random_state": 42,
}

# A list holds the SSE values for each k
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(df_pca)
    sse.append(kmeans.inertia_)
```

```python
 plt.style.use("fivethirtyeight")
 plt.plot(range(1, 11), sse)
 plt.xticks(range(1, 11))
 plt.xlabel("Number of Clusters")
 plt.ylabel("SSE")
 plt.show()
```

```python
kl = KneeLocator(
    range(1, 11), sse, curve="convex", direction="decreasing"
)

print('We have identify the the elbow of the curve: ', kl.elbow)
```

```python
# A list holds the silhouette coefficients for each k
silhouette_coefficients = []

# Notice you start at 2 clusters for silhouette coefficient
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(df_pca)
    score = silhouette_score(df_pca, kmeans.labels_)
    silhouette_coefficients.append(score)
```

```python
plt.style.use("fivethirtyeight")
plt.plot(range(2, 11), silhouette_coefficients)
plt.xticks(range(2, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()
```

Based on this graph we can consider that 2 is the best number of cluster because we have the maximum coefficient.
